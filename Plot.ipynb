```python
import pandas as pd
import matplotlib.pyplot as plt

DIR = "E:\\Projects\\Neural Net\\Galaxy Zoo\\Classification\\Results\\EfficientNetB0_94.15\\"
```

```python
history = pd.read_csv(DIR + "training_efficientnetB0_TPU_FC64.csv")
```

```python
history.dropna(inplace=True)
```

```python
history.columns
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 40,20

params = {'legend.fontsize': 40,
          'legend.handlelength': 4}
rcParams.update(params)
plt.rc('axes', labelsize=32)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=22)    # fontsize of the tick labels
plt.rc('ytick', labelsize=22)    # fontsize of the tick labels


plt.figure()
plt.plot(history['epoch'], history['loss'], label='Training Loss', linewidth = 5)
plt.plot(
    history['epoch'], history['val_loss'], label='Validation', linestyle='--',dashes=(5, 5), linewidth = 5)
plt.xlabel("Epochs")
plt.ylabel("RMSE")
plt.legend()
plt.savefig(DIR + 'Loss_B0.png')
plt.show()
```

```python
plt.figure()
plt.plot(history['epoch'], history['accuracy'], label='Training Accuracy', linewidth=5)
plt.plot(
    history['epoch'], history['val_accuracy'], label='Validation Accuracy', linestyle='--',dashes=(5, 5), linewidth=5)
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.savefig(DIR + 'Acc_B0.png')
plt.show()
```

```python
plt.figure()
plt.plot(history['epoch'], history['lr'], label='Learning Rate', linewidth=5)
plt.xlabel("Epochs")
plt.ylabel("Learning Rate")
plt.legend()
plt.savefig(DIR + 'Lr.png')
plt.show()
```
